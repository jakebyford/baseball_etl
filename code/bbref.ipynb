Note the list comprehensions. Scraping is an area comprehensions really become valuable. Now we have to do every row. HAving to do something a bunch of times should make you think about putting it in a function. So, let's make a function that'll parse some row (a BeautifulSoup tr tag)

```python
def parse_row(row):
    """
    Take in a tr tag and get data out of it in the form of a list of strings.
    """
    return [str(x.string) for x in row.find_all('td')]
```

```python
list_of_parsed_rows = [parse_row(row) for row in rows[2:-2]]
df = DataFrame(list_of_parsed_rows)
df
```

```python
# It doesn't have column names. Recall row number 1 was our header row
df.columns = parse_row(rows[1])
```

There we go, we've built our first webscraper!
Now to extend it to any team, let's put it in a function that takes a team, and does all the scraping inside it. Note we're using our HEADERS variable from above when calling requests.get

```python
def scrape_opening_day_single_team(team):
    team_response = requests.get(f'https://www.baseball-almanac.com/opening_day/odschedule.php?t={team}', headers = HEADERS)
    team_soup = Soup(team_response.text)
    
    tables = team_soup.find_all('table')
    
    team_table = tables[0]
    rows = team_table.find_all('tr')
    
    list_of_parsed_rows = [parse_row(row) for row in rows[2:-2]]
    df = DataFrame(list_of_parsed_rows)
    df.columns = parse_row(rows[1])
    
    # let's add in team as a column
    df['team'] = team
    
    return df
```

```python
# Let's try it
df_mil = scrape_opening_day_single_team('MIL')
df_mil.head()
```

```python
df_nyy = scrape_opening_day_single_team('NYA')
df_nyy
```

```python
# Going back to the main page and seeing if we can get a 'list of teams' to run through our function
mlb_response = requests.get(f'https://www.baseball-almanac.com/opening_day/opening_day.shtml', headers = HEADERS)
mlb_soup = Soup(mlb_response.text)
```

```python
mlb_soup
```

```python
# Let's find all the links
all_link_tags = mlb_soup.find_all('a')
all_link_tags
```

```python
# Looking through these, we can see we want the ones of the form
# 'team Opening Day Games'

# Let's use a list comprehension to find only those
all_opening_day_link_tags = [x for x in all_link_tags if
                            str(x.string).endswith('Opening Day Games')]
all_opening_day_link_tags
```

```python
# Finally, we can use normal Python string methods to get our list of teams
teams = [str(x.get('href'))[-3:] for x in all_opening_day_link_tags]
teams
```

```python
# Now we can loop through each of these and use our function 'scrape_opening_day...' to get the data we want
# Remember, pd.concat sticks data together. Let's use it with 3 teams without overloading baseball almanacs servers

df = pd.concat([scrape_opening_day_single_team(team) for team in teams [:3]])
df
```
